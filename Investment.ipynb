## Sector Analysis 1

```python
# Creating a new column named primary_sector in master_frame by splitting category_list column and
# choosing the first item from it
master_frame['primary_sector'] = master_frame['category_list'].str.split('|').str.get(0)
```

```python
# Converting the primary_sector of master_frame into lower case for effective analysis
master_frame['primary_sector'] = master_frame['primary_sector'].str.lower()
master_frame
```

```python
# Importing mapping.csv file storing it as mapping
mapping = pd.read_csv("mapping.csv")
mapping
```

```python
# Code for inspection
print(mapping.info())
print(mapping.shape)
```

```python
# Copying mapping into mapping_cleaned
mapping_cleaned = mapping
```

```python
# Cleaning the column category_list and converting it into lowercase in mapping_cleaned dataframe
mapping_cleaned['category_list'] = mapping_cleaned.category_list.str.replace('0','na')
mapping_cleaned['category_list'] = mapping_cleaned.category_list.str.replace('2.na','2.0')
mapping_cleaned['category_list'] = mapping_cleaned.category_list.str.lower()
mapping_cleaned
```

```python
# Creating a data frame named mapping_melted with category list, main sector and values columns
mapping_melted = pd.melt(mapping_cleaned,id_vars=['category_list'],var_name='main_sector', value_name='values')
mapping_melted
```

##### Here the NaN is automatically dropped when master_frame is merged along with mapping_melted. We can see it in further steps

```python
# Now exactly mapping each category_list to its corresponding main_sector by dropping all the '0' values
mapping_melted.drop(mapping_melted[mapping_melted.values == 0].index, inplace=True)
mapping_melted
```

```python
# Code for inspection
mapping_melted.info()
mapping_melted.shape
```

```python
# Dropping values column from mapping_melted dataframe
mapping_melted = mapping_melted.drop(columns=['values'])
```

```python
mapping_melted
```

```python
# Merging the master_frame with mapping_melted and dropping category_list_y column from the merged frame
merged_sector_master_frame = master_frame.merge(mapping_melted, left_on='primary_sector', right_on='category_list', how='inner')
merged_sector_master_frame = merged_sector_master_frame.drop(['category_list_y'], axis = 1)
merged_sector_master_frame
```

```python
# Code for inspection
print(merged_sector_master_frame.info())
print(merged_sector_master_frame.shape)
```

```python
# Checking the null values after merging
merged_sector_master_frame.isnull().sum()
```

####  Code for a merged data frame with each primary sector mapped to its main sector has been created
